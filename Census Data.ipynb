# 1. Which districts have the highest and lowest population?

```python
#highest population distict #MAHARASHTRA         Thane
print(data[data.Population == data.Population.max()])

#lowest population district #ARUNACHAL PRADESH  Dibang Valley
print(data[data.Population == data.Population.min()])
```

# 2. Which district had the highest number of males and females?

```python
print(data[data.Male == data.Male.max()])
# MAHARASHTRA         Thane
print(data[data.Female == data.Female.max()])
```

# 3. Which district had the lowest number of males and females

```python
print(data[data.Male == data.Male.min()])
      
#ARUNACHAL PRADESH  Dibang Valley
print(data[data.Female == data.Female.min()])
```
